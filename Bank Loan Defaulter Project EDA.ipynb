```python
Title:Bank Loan Defaulter Analysis Using EDA(Exploratory Data Analysis)
Introduction: In this project, I have conducted an indepth analysis of a dataset containing inforamtion about bank customers to predict loan defaulters.
The project aims to analyze and predict the likelihood of loan defaulters using Exploratory Data Analysis (EDA) techniques in Python.
The analysis focuses on understanding the key factors influencing loan default and building a predictive model to help financial institutions minimize the risk associated with lending.
Tools Used:
Python programming language
Pandas for data manipulation
Matplotlib and Seaborn for data visualization.
Target:
The primary goal of this project was to explore the relationships between different variables in the loan applicant dataset and identify any significant correlations.
This analysis aimed to provide insights into the factors that may influence the likelihood of loan default.
```

```python
Basic informations about the Dataset:
The DataFrame contains  100000 entries and 24 columns 
There are three data types present in the dataframe: int64, float64, and object.
The columns contain the following information:
SK_ID_CURR: Unique identifier for each loan applicant (int64)
TARGET: Binary indicator (0 or 1) of whether the loan applicant defaulted (int64)
NAME_CONTRACT_TYPE: Type of loan contract (object)
GENDER: Gender of the loan applicant (object)
Car: Indicator of whether the applicant owns a car (object)
House: Indicator of whether the applicant owns a house (object)
CNT_CHILDREN: Number of children the applicant has (int64)
AMT_INCOME_TOTAL: Total income of the applicant (float64)
AMT_CREDIT: Amount of credit requested by the applicant (float64)
AMT_GOODS_PRICE: Price of the goods for which the loan is requested (float64)
NAME_TYPE_SUITE: Who was accompanying the applicant when applying for the loan (object)
NAME_INCOME_TYPE: Type of income the applicant has (object)
NAME_EDUCATION_TYPE: Level of education of the applicant (object)
NAME_FAMILY_STATUS: Family status of the applicant (object)
DAYS_EMPLOYED: Number of days employed (int64)
MOBILE, WORK_PHONE, HOME_PHONE, MOBILE_REACHABLE, FLAG_EMAIL: Various indicators related to contact information (int64)
OCCUPATION_TYPE: Type of occupation of the applicant (object)
CNT_FAM_MEMBERS: Number of family members of the applicant (float64)
APPLICATION_DAY: Day of the week when the loan application was submitted (object)
TOTAL_DOC_SUBMITTED: Total number of documents submitted by the applicant(int64)

```

```python
#null value treatment:
loan_data['AMT_GOODS_PRICE'].fillna(loan_data['AMT_GOODS_PRICE'].mode()[0],inplace=True)
loan_data['NAME_TYPE_SUITE'].fillna(loan_data['NAME_TYPE_SUITE'].mode()[0],inplace=True)
loan_data['OCCUPATION_TYPE'].fillna(loan_data['OCCUPATION_TYPE'].mode()[0],inplace=True)
loan_data['CNT_FAM_MEMBERS'].fillna(loan_data['CNT_FAM_MEMBERS'].mode()[0],inplace=True)
```

```python
loan_data.duplicated()
```

```python
cat=[]
num=[]
def data_type(df):
    for col in df.columns:
        if df[col].nunique()<=15:
            cat.append(col)
        else:
            num.append(col)

data_type(loan_data)      
```

```python
def mode_cat(cat):
    for col in cat:
        print(f'{col},Mode={loan_data[col].mode()[0]},Nunique {loan_data[col].nunique()} , Missing_Val {loan_data[col].isnull().sum()}')
        
mode_cat(cat)   
```

```python
# Relationship between numerical and target variable using boxplot or scatter plots:
plt.figure(figsize=(8, 4))
sns.boxplot(x='TARGET',y='AMT_INCOME_TOTAL',data=loan_data)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='TARGET',y='AMT_CREDIT',data=loan_data)
plt.show()
```

```python
# Correlation Analysis:
correlation_matrix=loan_data[['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_GOODS_PRICE']].corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.show()
```

```python
conclusion:
The correlation analysis revealed that some variables in the dataset are highly correlated, with a correlation coefficient of 1.
This suggests a perfect linear relationship between these variables.
Understanding these correlations can help financial institutions make more informed decisions when assessing loan applications and managing the risk of default.
```
